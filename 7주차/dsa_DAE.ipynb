```python
# 노이즈를 추가할 열들 정의
columns_to_add_noise = df_filtered.columns.tolist()  # df_filtered의 모든 열에 노이즈를 추가할 경우

# 일부 열에만 노이즈를 추가하려면 예를 들어 이렇게 정의할 수 있음
# columns_to_add_noise = ['column1', 'column2', 'column3']  # 원하는 열을 지정

# df_noised는 df_filtered의 복사본이므로 원본 데이터는 영향을 받지 않음
df_noised = df_filtered.copy()

# 노이즈 추가
noise_factor = np.random.uniform(0, 1)

for column in columns_to_add_noise:
    df_noised[column] += noise_factor * np.random.normal(loc=0.0, scale=1.0, size=df_noised[column].shape)

# 노이즈가 추가된 데이터 확인
df_noised.head()
```

```python
# normal = df_filtered[(df_filtered.activity == 0)

normal = df_filtered[(df_noised.activity == 0)]

# activity가 1인 데이터를 선택
abnormal = df_filtered[df_noised.activity == 1]

# normal 데이터프레임의 모양 확인
normal.shape
abnormal.shape
```

```python
import numpy as np

# 노이즈가 추가된 데이터와 원본 데이터를 준비
noise_train = df_noised.drop(['activity'], axis=1)
not_noise_train = df_filtered.drop(['activity'], axis=1)

# 'activity' 컬럼에서 고유한 값과 빈도 계산
unique_values, counts = np.unique(df['activity'], return_counts=True)

# 결과 출력
print(dict(zip(unique_values, counts)))
```

```python
from tensorflow.keras.optimizers import Adam

input_dim = noise_train.shape[1]
encoding_dim = 16  # 인코딩 차원을 설정 (임의로 설정한 값, 조정 가능)

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
encoded = Dense(int(encoding_dim / 2), activation='relu')(encoded)
decoded = Dense(int(encoding_dim /2), activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
autoencoder.summary()
```

```python
autoencoder.fit(noise_train, not_noise_train, epochs=50, batch_size=64, shuffle=True, verbose=1)
```

```python
from sklearn.model_selection import train_test_split

X_noise_train, X_noise_test = train_test_split(df_noised, test_size=0.25, random_state=20)

X_noise_train.head()
```

```python
# normal feature만 가지는 X_train
# 0: normal, 1: abnormal
X_noise_train = X_noise_train[X_noise_train.activity == 0]

y_train = X_noise_train['activity']
y_test = X_noise_test['activity']

X_noise_train = X_noise_train.drop(['activity'], axis=1)
X_noise_test = X_noise_test.drop(['activity'], axis=1)

X_noise_train.head()
```

```python
denoised_X_test = autoencoder.predict(X_noise_test)
```

```python
# 오토인코더 모델

input_dim = denoised_X_test.shape[1]
encoding_dim = 16  # 인코딩 차원을 설정 (임의로 설정한 값, 조정 가능)

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
encoded = Dense(int(encoding_dim / 2), activation='relu')(encoded)
decoded = Dense(int(encoding_dim /2), activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
autoencoder.summary()
```

```python
history = autoencoder.fit(denoised_X_test, denoised_X_test, epochs=25, batch_size=256, shuffle=True, validation_split = 0.1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right');
```
